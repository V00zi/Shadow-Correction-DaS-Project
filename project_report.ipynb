## Introduction

The traditional perspective on digital piracy views it through a strictly punitive lens: piracy is detrimental because it perpetuates the illegal distribution of digital media (games, movies, series, software) and incurs direct financial harm to the market. Under this conventional misconception, anti-piracy efforts are driven by the belief that eliminating pirate channels will naturally convert pirates into paying customers.

However, this study proposes a paradigm shift, moving away from identifying bad consumer market practices to analyzing how piracy acts as a corrective force. Our core hypothesis is that **piracy is a predictable market correction that occurs whenever an existing market trend works against consumers' satisfaction**. Historically, the digital media landscape benefited from a centralized, convenient model (e.g., the early days of Steam and Netflix). Today, that landscape has shifted to a highly fragmented and expensive one, breeding new grounds for market gaps that did not previously exist. This research aims to analyze why these gaps exist and how piracy effectively corrects them.

## Project Objectives

To empirically test our hypothesis, this study is structured around three primary objectives:

1. **The Economic Barrier:** To quantify the correlation between Purchasing Power and piracy search volumes across different geographical regions.


2. **The Fragmentation Barrier:** To determine if the increased fragmentation of the streaming market positively correlates with piracy search traffic.


3. **The Preservation Barrier:** To evaluate piracy behavior following the removal (delisting) of digital media from storefronts.


## Literature Review

The research papers aided perfectly to align our project's core hypothesis: that digital piracy should be viewed not through a strictly punitive lens, but as a predictable market correction responding to economic, service, and accessibility barriers.

Brief overview of how each paper relates to our project's objectives:

#### 1. "Deliberate, Principled, Self-Interested Lawbreaking: The Ethics of Digital Piracy" (Breakey, 2018)

- This paper supports our project’s move away from viewing piracy as purely criminal behavior. It argues that piracy is often a deliberate and principled response to industry malfeasance, exorbitant pricing, or lack of access. It directly infers our "Economic Barrier" and "Preservation Barrier" concepts by showing that for many, especially in developing nations, piracy is a necessary tool to overcome wealth disparities and access culturally significant media that is otherwise unavailable.

#### 2. "Piracy as a Market Strategy for Video Games" (Tripathi et al., 2025)
- This paper is foundational for our analysis of the "Service and Convenience gaps". It argues that in emerging economies, rampant piracy is essentially a "global pricing problem" because legitimate games are priced like luxury items relative to local incomes. It highlights Valve CEO Gabe Newell's insight that piracy is a service problem, proving that providing localized pricing and convenient digital access is far more effective at stopping piracy than strict enforcement. This directly infers our hypothesis that piracy bridges the gap when market trends work against consumer convenience and affordability.

#### 3. "The Decline of Online Piracy: How Markets - Not Enforcement - Drive Down Copyright Infringement" (Quintais & Poort, 2019)
- This study provides the empirical backing for our overarching claim that market dynamics dictate piracy rates. It concludes that the decline in online piracy is driven primarily by the increasing availability of affordable and convenient legal content, rather than by legal enforcement or punitive measures. It directly supports our "Fragmentation Barrier" objective, reinforcing the idea that when the market provides a centralized, affordable, and convenient model, users are willing to pay, but when the market fractures or becomes too expensive, consumers turn to piracy.

- This paper provides the methodological framework for testing our "Economic Barrier". By analyzing the correlation between regional Steam game prices, GDP per capita, and video game penetration rates, it demonstrates how companies use price discrimination based on local purchasing power. our project mirrors this methodology by utilizing World Bank data (GDP) and SteamDB to test the correlation between regional economics and piracy search volumes.

### 1. Testing the Economic Barrier

To investigate if the worth of digital media is vastly different and unfair depending on geographical regions, we tested the correlation between regional economics and piracy.

* **Method:** Pearson Correlation Coefficient ($r$) and Linear Regression.

* **Success Criteria:** A strong positive correlation ($r > 0.7$) would statistically prove that poverty drives piracy, refuting the argument that it is purely a cultural/criminal issue.





### 2. Testing the Fragmentation Barrier

To determine if declining market concentration (increasing competition) drives users away from legal options.

* **Method:** Herfindahl-Hirschman Index (HHI) for market concentration (scaled 0 to 10,000) and Cumulative Cost Threshold Analysis.


* **Application:** Plotting Piracy Volume against the combined 'Cost of the Top 3 Bundles'. We hypothesize that piracy triggers specifically when this cost exceeds 2% of the average monthly income in a region.


* **Success Criteria:** Identifying a specific 'Price Cliff' where casual users mass-migrate to piracy, proving fragmentation is an economic friction problem.


### 3. Testing the Preservation Barrier

To understand if piracy is used as a means of preservation for media that is going extinct (delisted).

* **Method:** Interrupted Time Series (ITS) Analysis / Event Study.


* **Application:** Measuring the mean search volume for sampled games 4 months before being delisted versus 4 months after.


* **Success Criteria:** A statistically significant increase in search volume post-removal would validate the "Streisand Effect," proving piracy acts as an archival mechanism.

### 1.2 Statistical Validation

To test the visual findings, both parametric and non-parametric statistical methods were applied.


```python
# We add a constant because statsmodels doesn't include the intercept (beta_0) by default
X = sm.add_constant(df_sorted['Hourly Wages (estimated)'])
y = df_sorted['Piracy Searches']

# Fit the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

print(model.summary())
```


**A. Ordinary Least Squares (OLS) Regression**
A simple linear regression was fitted to model Piracy Searches against Hourly Wages.

* **Coefficient:** -0.1884
* **P-value:** 0.141
* **R-squared:** 0.021

*Interpretation:* The negative coefficient aligns with our initial hypothesis (wages negatively affect piracy). However, a p-value of 0.141 is strictly greater than the standard **alpha level = 0.05**.

 Therefore, we *fail to reject the null hypothesis* (H0: The Hourly Wages and Piracy volumes are negatively correlated). Furthermore, an R-squared of 0.021 indicates that hourly wages explain only roughly 2% of the variance in piracy searches.


```python
# Calculate Spearman's Rho and p-value
rho, p_value = stats.spearmanr(df_sorted['Hourly Wages (estimated)'],
                               df_sorted['Piracy Searches'],
                               nan_policy='omit')

print(f"Spearman's Rank Correlation (Rho): {rho:.4f}")
print(f"P-value: {p_value:.4e}")
```

```python
correlation = df_sorted[['Hourly Wages (estimated)', 'Piracy Searches']].corr(method='spearman')
print(correlation)
```


**B. Spearman's Rank Correlation**
Because economic data is often non-linear and susceptible to extreme outliers, Spearman's Rank Correlation was utilized as a robust, non-parametric alternative to check for a monotonic relationship.

# Parameter 2: The Fragmentation Barrier


**Objective:** To determine if the increased fragmentation of the streaming market positively correlates with global video piracy. ### The Shadow Correction Hypothesis

The traditional view treats piracy as static criminal behavior. Our hypothesis proposes a dynamic market correction model: **when legal services fragment beyond consumer tolerance, piracy doesn't simply increase, it evolves into a professionalized infrastructure that mirrors the convenience once offered by monopolistic Netflix, but circumvents fragmentation costs.**

This creates three distinct evolutionary phases:
1. **Phase 1 (2011-2016): The Golden Age** - Good service kills piracy
2. **Phase 2 (2018-2023): The Fragmentation Paradox** - Search behavior declines  
3. **Phase 3 (2018-2023): The Shadow Market Revelation** - Actual piracy surges

### 2.2 Calculating the Herfindahl-Hirschman Index (HHI)

The HHI quantifies market concentration as the sum of squared market shares:

$$\text{HHI} = \sum_{i=1}^{n} s_i^2 \times 10000$$

Where $s_i$ is the market share of platform $i$ (expressed as a percentage).

**DOJ/FTC Classification:**
- **HHI > 2,500:** Highly concentrated (monopoly/oligopoly)
- **HHI 1,500-2,500:** Moderate concentration  
- **HHI < 1,500:** Competitive market

**Fragmentation Index** = 10,000 - HHI (for inverse interpretation: higher = more fragmentation)


### 2.2.1 Operationalizing the Fragmentation Barrier

The **fragmentation barrier** is not merely the count of streaming platforms, but rather the compounded user friction created by market fragmentation. We operationalize this barrier through three interrelated dimensions:

1. **Market Concentration (HHI):** Quantifies platform distribution and competitive intensity
2. **Service Exclusivity:** Content dispersion across multiple subscription-required platforms  
3. **Transaction Costs:** Cumulative subscription fees, search friction, and decision fatigue

**Fragmentation Barrier Proxy:**

We proxy the fragmentation barrier using declining HHI (increasing platform competition) combined with rising exclusive content distribution, which creates structural friction that pushes consumers toward piracy as a "one-stop solution."

Mathematically, higher fragmentation barrier corresponds to:
- Lower HHI (more dispersed market)
- Higher platform count with exclusive catalogs
- Increased total subscription cost to access equivalent content

This operationalization allows us to test whether market fragmentation correlates with piracy resurgence, consistent with the service-friction hypothesis.

### 2.3 Phase 1: The Golden Age (2011-2016)
**Monopolistic Convenience and Piracy Suppression**

Gabe Newell's famous quote from 2011 proposed that piracy exists primarily because legal services are inconvenient, not because people refuse to pay. The early Netflix era (2011-2016) provided an empirical test of this hypothesis.

**US Market Data (Sandvine):**
- **2011 Q4:** Netflix commanded 97.6% of streaming market share (HHI = 9,533 - near monopoly)
- **2016 Q2:** Netflix share declined to 83.5% as Amazon entered (HHI = 7,140 - still highly concentrated)

**Concurrent Piracy Behavior:**
- **BitTorrent traffic:** Declined from 18.87% of total North American traffic (2011) to below 10% (2015)
- **Google Trends searches (US):** Piracy-related searches declined as Netflix penetration increased

**Critical Insight:** When a single, convenient, affordable service offered comprehensive content, consumers gladly paid instead of pirating. This established the baseline that **consumers prefer paying for convenience when service quality is high.**

**Temporal Precedence:** This phase demonstrates that US market fragmentation began in 2011-2012 (Amazon's entry), a full **7 years before** reliable global piracy measurement (MUSO) became available in 2018. This temporal sequence suggests fragmentation preceded the global piracy surge documented in our primary analysis.

### 2.4 Phase 2: The Fragmentation Paradox (2018-2023)
**Search Behavior Decline Amid Market Fragmentation**

**The Contradiction:** As the global streaming market fragmented dramatically (Disney+, Apple TV+, HBO Max all launched), Google Trends data showed an unexpected pattern: worldwide piracy-related searches continued to **decline**.

**Google Trends Data (Worldwide, 2018-2023):**
- Composite search index: "putlocker + watch movies free + watch series free + 123movies + fmovies"
- Search interest showed declining trend as fragmentation increased

**Market Fragmentation Data (Parrot Analytics - Worldwide):**
- **2018:** HHI = 5,276 (concentrated)
- **2023:** HHI = 2,101 (moderately concentrated)
- **Overall decline: 60.2%**

**The Apparent Contradiction:**

Based on preliminary correlation analysis between HHI and Google Trends search interest for piracy terms, we observe that as market concentration decreased (fragmentation increased), search interest also declined. This suggests:
- Higher market concentration (monopoly) → MORE piracy searches
- Lower market concentration (fragmentation) → FEWER piracy searches

**This is the opposite of what Shadow Correction would predict!** If fragmentation drives piracy, why are people searching for it less?

### 2.5 Phase 3: The Shadow Market Revelation (2018-2023)
**Fragmentation-Induced Informal Market Professionalization**

**The Resolution:** MUSO actual traffic data reveals what Google Trends cannot see.

**MUSO Global Piracy Data (Worldwide):**
- **2018:** 127.5 billion video piracy visits
- **2023:** 141.0 billion visits
- **Overall increase: +10.6%**

**The Critical Discovery:**

While Google searches for piracy declined globally, **actual piracy visits increased 10.6%**. This reveals a fundamental behavioral evolution:

**How Pirates Evolved (2011 → 2023):**

| Discovery Method | 2011 Era | 2023 Era |
|-----------------|----------|----------|
| **Search-Dependent** | Google "torrent Game of Thrones" | Obsolete |
| **Pre-Configured Apps** | Rare, clunky | Stremio, Popcorn Time, Kodi |
| **Social Sharing** | Forum links | Discord servers, Telegram channels, Reddit megathreads |
| **IPTV Services** | Non-existent | Pre-loaded boxes sold on Amazon |
| **Direct Bookmarks** | Limited | Curated lists, RSS feeds |
| **Search Engine Reliance** | Essential | Irrelevant |

**The Shadow Market Matured:** By 2023, piracy infrastructure had evolved into a sophisticated, user-friendly ecosystem that perfectly replicates Netflix's convenience but circumvents fragmentation costs. Users no longer "search" for piracy, they access it directly through established channels.

**The Evolution Explained:**

1. **2011-2016:** Piracy = technical hassle (torrents, seeders, codecs). Netflix wins by being easier.
2. **2018-2023:** Legal market fragments. Shadow market professionalizes. Apps like Stremio offer one-click access to ALL content from ALL platforms, free.

### 3.3 Normality Diagnostics (Q-Q Plot)

To verify the hypothesis that the Time Delta follows a Normal Distribution, a Quantile-Quantile (Q-Q) plot was generated.



```python
# Data Cleaning & Transformation and plotting the Q-Q Plot
# Convert to datetime and sort/filter as per your requirement
par2['Date'] = pd.to_datetime(par2['Date'])
par2['Date Delisted'] = pd.to_datetime(par2['Date Delisted'])

par2_unique = par2.loc[par2.groupby('Keyword')['Value'].idxmax()].copy()

# Generating the Q-Q Plot
plt.figure(figsize=(8, 6))

stats.probplot(par2['Difference'].dropna(), dist="norm", plot=plt)


plt.title('Probability Plot: Q-Q for Days to Delist', fontsize=14, fontweight='bold')
plt.xlabel('Theoretical Quantiles (Normal)', fontweight='bold')
plt.ylabel('Ordered Values (Actual Days)', fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```


**Interpretation of the Q-Q Plot:**

The Quantile-Quantile (Q-Q) plot is the gold standard for verifying if a dataset follows a specific distribution (in this case, if $\Delta T$ follows a **Normal Distribution** $\mathcal{N}(\mu, \sigma^2)$)



1. **Linearity:** The majority of the empirical quantiles (blue dots) follow the theoretical red line closely between $-1$ and $+1$ standard deviations. This confirms that the "middle" of your data is **approximately normally distributed**.


2. **Tail Behavior (The "S-Curve"):**
* **Lower Left:** The dots curve *above* the red line. This indicates that the smallest values in your dataset are "less negative" than what a normal distribution would predict. This suggests a **lower bound** (data is truncated on the left).
* **Upper Right:** The dots curve *below* the red line. This indicates that the largest values are "less positive" than expected.
The data is more "tightly packed" than a perfect bell curve, with fewer extreme outliers than expected.


3.  **Conclusion** While the data is not *perfectly* normal due to the light tails and slight skew, it is **sufficiently normal** to justify using parametric statistical tests (like t-tests or ANOVA) for further analysis.

### 3.4 Data Integrity and Uncertainty Analysis

While the distributional analysis suggests a tendency toward normality, it is critical to account for temporal uncertainty in the `Date Delisted` attribute. Due to the lack of day-specific granularity in the source metadata, all delisting events were normalized to the **1st of the respective month**.

#### 3.4.1 Systematic Error Margin

This assumption introduces a systematic bias where the recorded `Difference` represents the **minimum possible delta** (the "Earliest Potential Delisting"). The actual delisting date could have occurred at any point within that month ($1 \dots N$).

We define the error margin for each observation $i$ as:


$$\text{Difference}_{actual} = \text{Difference}_{recorded} + \epsilon_i$$


Where the error term $\epsilon_i$ is constrained by the month's length ($N$):


$$0 \le \epsilon_i \le (N - 1)$$

#### 3.4.2 Impact on Distributional Findings

To account for this inconsistency in the project report, we apply a **Midpoint Adjustment Strategy**:

* **Average Error:** For a standard 30-day month, the uncertainty is $\pm 15$ days.
* **Shift in Central Tendency:** If delistings are distributed randomly throughout the month, our current mean of **-8.0 days** would likely shift toward a more positive value (approximately **+7.0 days**), suggesting that search interest peaks slightly *after* the actual delisting occurs, rather than before.

#### 3.4.3 Conclusion on Preservation

Despite this $\pm 15$-day "blurring" of the data, the **normality of the distribution remains robust**. The error is distributed uniformly across the dataset and does not favor specific keywords, meaning the overall "S-Curve" observed in the Q-Q plot remains a valid representation of the market's reaction to digital delisting.
